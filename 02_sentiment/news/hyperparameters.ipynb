```python
import os
import json
import ast
```

```python
files = [x for x in os.listdir("cv_results_v1") if x.endswith(".json")]

hyp_v1 = {}
for file in files:
    with open(os.path.join("cv_results_v1", file), "r") as f:
        data = json.load(f)
        if file.split()[0] not in hyp_v1:
            hyp_v1[file.split()[0]] = list(data.keys())
        else:
            hyp_v1[file.split()[0]] = hyp_v1[file.split()[0]] + list(data.keys())
```

```python
hyp_v2 = {}

files = [x for x in os.listdir("cv_results_v2/bertemb") if x.endswith(".json")]
hyp_v2["bertemb"] = []
for file in files:
    dict_str = file.split("}_")[0] + "}"
    dict_str = dict_str.replace("'-", "':")
    result = ast.literal_eval(dict_str)

    hyp_v2["bertemb"].append(result)

files = [x for x in os.listdir("cv_results_v2/sectoremb") if x.endswith(".json")]
hyp_v2["sectoremb"] = []
for file in files:
    dict_str = file.split("}_")[0] + "}"
    dict_str = dict_str.replace("'-", "':")
    result = ast.literal_eval(dict_str)

    hyp_v2["sectoremb"].append(result)

files = [x for x in os.listdir("cv_results_v2/tfidfemb") if x.endswith(".json")]
hyp_v2["tfidfemb"] = []
for file in files:
    dict_str = file.split("}_")[0] + "}"
    dict_str = dict_str.replace("'-", "':")
    result = ast.literal_eval(dict_str)

    hyp_v2["tfidfemb"].append(result)
```

```python
trials1 = {sector: len(hyps) for sector, hyps in hyp_v1.items()}
trials2 = {sector: len(hyps) for sector, hyps in hyp_v2.items()}
trials1, trials2
```

```python
list_of_hyp = {}
for model, hyp_list in hyp_v2.items():
    if model not in list_of_hyp:
        list_of_hyp[model] = {}
    
    for hyp in hyp_list:
        for hyp_name, value in hyp.items():
            if hyp_name not in list_of_hyp[model]:
                list_of_hyp[model][hyp_name] = set()
            list_of_hyp[model][hyp_name].add(value)
```

```python
list_of_hyp_v1 = {}
for model, hyp_list in hyp_v1.items():
    if model not in list_of_hyp_v1:
        list_of_hyp_v1[model] = {}

        for hyp in hyp_list:
            hyp_set = hyp.split(" | ")

            for i, parm in enumerate(hyp_set):
                if i not in list_of_hyp_v1[model]:
                    list_of_hyp_v1[model][i] = set()
                
                list_of_hyp_v1[model][i].add(parm)
```
